## Data Cleaning

### 1. Remove dollars phrase from price column

```python
df['price']=df['price'].str.strip('dollars') #df['price'].str.replace('dollars','')
```

```python
df
```

### 2. Convert last review column as datetime column and price column as numeric column

```python
df['last_review']=pd.to_datetime(df['last_review'])
```

```python
df['price']=pd.to_numeric(df['price'])
```

```python
df.info()
```

### 3. Make the room_type column to lower case to have uniformity in the data

```python
df['room_type']=df['room_type'].str.lower()
df['room_type']
```

### 4. Check for null values

```python
df.isna().sum()
```

```python
df
```

### 5. Add Month column

```python
df['month']=df['last_review'].dt.month
df
```

### 6. Add price_per_month column

```python
df['price_per_month']=df['price']*30.5
df
```

```python
df.to_csv('Final_Combined_data.csv',index=False)
```

```python
def func(a):
    if a>0 and a<70:
        return '$0-69'
    elif a>=70 and a<=175:
        return '$70-175'
    elif a>175 and a<=350:
        return '$176-350'
    else:
        return '>$350'

df['price_range']=df['price'].apply(func)

def func(a):
    if a=='$0-69':
        return 'Budget'
    elif a=='$70-175':
        return 'Average'
    elif a=='$176-350':
        return 'Expensive'
    else:
        return 'Extravagant'

df['label']=df['price_range'].apply(func)
```
